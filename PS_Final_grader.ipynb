#### Packages and File Paths

```python
import os
import pandas as pd
import numpy as np
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

######################################
#File paths
######################################
PATH  = '/content/drive/MyDrive/Grading'

# solution_file = PATH  + '/Final Spreadsheet Assignment - SOLUTION.xlsx'
solution_file = 'https://www.dropbox.com/scl/fi/m68ntsd892pas74lkeyld/Final-Spreadsheet-Assignment-SOLUTION.xlsx?rlkey=4hxbxxejgdu8bwljvlbccj35k&st=0btcyeoj&dl=1'
ps_path       = PATH + '/ps_final'
output_file   = PATH + '/grades_ps_final.xlsx'

# auxiliary function
def percent_deviation(student,correct,fraction):
    pct_dev = np.abs(student - correct)/correct
    return (pct_dev < fraction)
```

#### Read in solutions

```python
######################################
#Read in solutions
######################################

cells_to_pull1 = ['C17', 'C20', 'C21', 'C22', 'C25', 'C26', 'C27', 'C28', 'C29', 'B30', 'C34', 'C35', 'C36', 'C39', 'C40', 'C41', 'C42', 'B44', 'C49', 'C50', 'C51', 'C52', 'C55', 'C56', 'B56']
cells_to_pull = cells_to_pull1
solution1 = pd.DataFrame(index=cells_to_pull, columns=['q1'])
for i in enumerate(cells_to_pull):  
    col_to_pull =str(cells_to_pull[i[0]][0])
    row_to_pull =int(cells_to_pull[i[0]][1:])-1       
    solution1.iloc[i[0],0]=pd.read_excel(solution_file,sheet_name='Q1', usecols=col_to_pull,skiprows=row_to_pull,nrows=1,header=None).iloc[0,0]

cells_to_pull2 = ['B49', 'B45','C48','D48','E48','F48','G48', \
                 'C46','D46','E46','F46','G46','C38','D38','E38','F38','G38', \
                 'C19','C20','C21','C22','C23','C24','C25','C26','C27','C28','C29','C32','C33','C35','C36','C37','C40','C41','C42','C43','C44','C45', \
                 'C53','C63','C58','C59', 'B63']    
cells_to_pull = cells_to_pull2
solution2 = pd.DataFrame(index=cells_to_pull, columns=['q2'])
for i in enumerate(cells_to_pull):  
    col_to_pull =str(cells_to_pull[i[0]][0])
    row_to_pull =int(cells_to_pull[i[0]][1:])-1       
    solution2.iloc[i[0],0]=pd.read_excel(solution_file,sheet_name='Q2', usecols=col_to_pull,skiprows=row_to_pull,nrows=1,header=None).iloc[0,0]

cells_to_pull3 = ['C10','C15','C20','B12','B17','B22','B21']
cells_to_pull = cells_to_pull3
solution3 = pd.DataFrame(index=cells_to_pull, columns=['q3'])
for i in enumerate(cells_to_pull):  
    col_to_pull =str(cells_to_pull[i[0]][0])
    row_to_pull =int(cells_to_pull[i[0]][1:])-1       
    solution3.iloc[i[0],0]=pd.read_excel(solution_file,sheet_name='Q3', usecols=col_to_pull,skiprows=row_to_pull,nrows=1,header=None).iloc[0,0]

```

#### Read in student solutions 

- should be downloaded and unzipped from canvas and place in Grading/ps_final folder of google drive

```python
######################################
# Read in student submissions
# Downloaded and unzipped from canvas    
######################################    
#Pull file names for the particular folder path
file_list = os.listdir(ps_path)
os.chdir(ps_path)
zeros = np.zeros(len(file_list), dtype=[('q1a','f8'),('q1b','f8'),('q1d','f8'),('q1f','f8'),('q1c_expl','str'),('q1e_expl','str'),('q1_total','f8'), \
                                        ('q2_repurchase','f8'), ('q2_bs_tie','f8'),('q2_IS_accts','f8'),('q2_BS_accts','f8'),('q2_datatable','f8'),('q2_total','f8'), \
                                        ('q3a','f8'), ('q3b','f8'), ('q3c','f8'),('q3a_expl','str'),('q3b_expl','str'),('q3c_expl','str'), ('q3_total','f8'), \
                                        ('q1_err','f8'),('q2_err','f8'),('q3_err','f8')])
deductions = pd.DataFrame(zeros)    
```

```python
for f in enumerate(file_list):
    print(f[1])    
    
    # Question 1 - Max deduction is 32 points:
    cells_to_pull = cells_to_pull1
    ans1 = pd.DataFrame(index=cells_to_pull, columns=['q1'])
    try:
        for i in enumerate(cells_to_pull):
    
            # print(str(i[0])+' is '+str(i[1]))  
            # print(str(cells_to_pull[i[0]][1:]))
            
            col_to_pull =str(cells_to_pull[i[0]][0])
            row_to_pull =int(cells_to_pull[i[0]][1:])-1 
            # print(str(row_to_pull))
                
            ans1.iloc[i[0],0]=pd.read_excel(f[1],sheet_name='Q1', usecols=col_to_pull,skiprows=row_to_pull,nrows=1,header=None).iloc[0,0]
            # ans.iloc[i[0],0]=pd.read_excel(file_list[0],sheet_name='Part 1', usecols='C',skiprows=14,nrows=1,header=None).iloc[0,0]

    except:
        print('Q1 Read error')
        deductions.loc[f[0],'q1_err']=10  
  
    
    try:    
        if deductions.loc[f[0],'q1_err']==0:
        
            #Check for row insertion
            if ans1.loc['B56'][0]!=solution1.loc['B56'][0]:
                print('Q1 row problem')
                deductions.loc[f[0],'q1_err']=1         
            
            ##### Q1a
             #Correlation correct?;
            if not(round(ans1.loc['C17'][0],2)==round(solution1.loc['C17'][0],2)):
                    deductions.loc[f[0],'q1a'] = deductions.loc[f[0],'q1a']-5
                
                
            ##### Q1b                
            w1 = ans1.loc['C20'][0]
            w2 = ans1.loc['C21'][0]
            w3 = ans1.loc['C22'][0]
            #Frontier weights correct?;
            if not(round(ans1.loc['C20'][0],2)==round(solution1.loc['C20'][0],2)):
                    deductions.loc[f[0],'q1b'] = deductions.loc[f[0],'q1b']-5
    
            #Expected Return correct?
            if not(round(ans1.loc['C25'][0],2)==round(w1*0.12 + w2*0.05 + w3*0.15,2)):
                    deductions.loc[f[0],'q1b'] = deductions.loc[f[0],'q1b']-2
    
            #Standard deviation correct?
            if not(round(ans1.loc['C27'][0],2)==round(np.sqrt((w1**2)*0.22**2 + (w2**2)*0.10**2 + (w3**2)*0.18**2 + 2*w1*w2*0.0110 + 2*w1*w3*0.0198 + 2*w2*w3*0.009) ,2)):
                    deductions.loc[f[0],'q1b'] = deductions.loc[f[0],'q1b']-2
    
            #Sharpe ratio correct?
            if not(round(ans1.loc['C28'][0],2)==round( (ans1.loc['C25'][0]-0.02)/ans1.loc['C27'][0]  ,2)):
                    deductions.loc[f[0],'q1b'] = deductions.loc[f[0],'q1b']-2
            deductions.loc[f[0],'q1c_expl']=ans1.loc['B30'][0]
    


            ##### Q1c                
            w1 = ans1.loc['C34'][0]
            w2 = ans1.loc['C35'][0]
            w3 = ans1.loc['C36'][0]
            #Tangency weights correct?;
            if not(round(ans1.loc['C34'][0],2)==round(solution1.loc['C34'][0],2)):
                    deductions.loc[f[0],'q1d'] = deductions.loc[f[0],'q1d']-5
    
            #Expected Return correct?
            if not(round(ans1.loc['C39'][0],2)==round(w1*0.12 + w2*0.05 + w3*0.15,2)):
                    deductions.loc[f[0],'q1d'] = deductions.loc[f[0],'q1d']-2
    
            #Standard deviation correct?
            if not(round(ans1.loc['C41'][0],2)==round(np.sqrt((w1**2)*0.22**2 + (w2**2)*0.10**2 + (w3**2)*0.18**2 + 2*w1*w2*0.0110 + 2*w1*w3*0.0198 + 2*w2*w3*0.009) ,2)):
                    deductions.loc[f[0],'q1d'] = deductions.loc[f[0],'q1d']-2
    
            #Sharpe ratio correct?
            if not(round(ans1.loc['C42'][0],2)==round( (ans1.loc['C39'][0]-0.02)/ans1.loc['C41'][0]  ,2)):
                    deductions.loc[f[0],'q1d'] = deductions.loc[f[0],'q1d']-2
            deductions.loc[f[0],'q1e_expl']=ans1.loc['B44'][0]
                  
            
            
            
            ##### Q1f     
            w_rf =ans1.loc['C49'][0]
            w1 = ans1.loc['C50'][0]
            w2 = ans1.loc['C51'][0]
            w3 = ans1.loc['C52'][0]
            #Risk-free weight correct?;
            if not(round(ans1.loc['C49'][0],2)==round(solution1.loc['C49'][0],2)):
                    deductions.loc[f[0],'q1f'] = deductions.loc[f[0],'q1f']-5
    
            #Expected Return correct?
            if not(round(ans1.loc['C55'][0],2)==round(w_rf*0.02 + w1*0.12 + w2*0.05 + w3*0.15,2)):
                    deductions.loc[f[0],'q1f'] = deductions.loc[f[0],'q1f']-2
    
            #Standard deviation correct?
            if not(round(ans1.loc['C56'][0],2)==round(np.sqrt((w1**2)*0.22**2 + (w2**2)*0.10**2 + (w3**2)*0.18**2 + 2*w1*w2*0.0110 + 2*w1*w3*0.0198 + 2*w2*w3*0.009) ,2)):
                    deductions.loc[f[0],'q1f'] = deductions.loc[f[0],'q1f']-2
    
            deductions.loc[f[0],'q1_total'] = sum(deductions.loc[f[0],'q1a':'q1f'])
    except:
        print('Q1 Grade error')
        deductions.loc[f[0],'q1_err']=1         
 
    
    
    
    
    
    
    
    
    
    
    # Question 2 - max deduction is 21 points:
    cells_to_pull = cells_to_pull2
    ans2 = pd.DataFrame(index=cells_to_pull, columns=['q2'])
    try:
        for i in enumerate(cells_to_pull):
            col_to_pull =str(cells_to_pull[i[0]][0])
            row_to_pull =int(cells_to_pull[i[0]][1:])-1        
            try: 
                ans2.iloc[i[0],0]=pd.read_excel(f[1],sheet_name='Q2', usecols=col_to_pull,skiprows=row_to_pull,nrows=1,header=None).iloc[0,0]  
            except ValueError:
                print('a first error')
                deductions.loc[f[0],'q2'] =-10
    except:
        print('Q2 Read error')
        deductions.loc[f[0],'q2_err']=1  
 

    try:    
        if deductions.loc[f[0],'q2_err']==0:
            #Average repurchase correct?
            if np.abs(round(ans2.loc['B49'][0],0))==np.abs(round(solution2.loc['B49'][0],0)):
                   deductions.loc[f[0],'q2_repurchase'] = deductions.loc[f[0],'q2_repurchase']-0 
                   
                   
            else:
                
                #Check repurchase numbers 
                if not(np.abs(round(ans2.loc['C45'][0],0) - round(ans2.loc['B45'][0],0))==np.abs(round(ans2.loc['C48'][0],0))):
                   deductions.loc[f[0],'q2_repurchase'] = deductions.loc[f[0],'q2_repurchase']-3 
                   
                   
                #Check balance sheet ties
                if not(round(ans2.loc['C46'][0],0)==round(ans2.loc['C38'][0],0) and \
                       round(ans2.loc['D46'][0],0)==round(ans2.loc['D38'][0],0) and \
                       round(ans2.loc['E46'][0],0)==round(ans2.loc['E38'][0],0) and \
                       round(ans2.loc['F46'][0],0)==round(ans2.loc['F38'][0],0) and \
                       round(ans2.loc['G46'][0],0)==round(ans2.loc['G38'][0],0) ):
                     deductions.loc[f[0],'q2_bs_tie'] = deductions.loc[f[0],'q2_bs_tie']-10               
                
                
                #Check first year income statement numbers
                if not(np.abs(round(ans2.loc['C19'][0],0))==np.abs(round(solution2.loc['C19'][0],0))):
                     deductions.loc[f[0],'q2_IS_accts'] = deductions.loc[f[0],'q2_IS_accts']-1                          
                if not(np.abs(round(ans2.loc['C20'][0],0))==np.abs(round(solution2.loc['C20'][0],0))):
                     deductions.loc[f[0],'q2_IS_accts'] = deductions.loc[f[0],'q2_IS_accts']-1   
                if not(np.abs(round(ans2.loc['C21'][0],0))==np.abs(round(solution2.loc['C21'][0],0))):
                     deductions.loc[f[0],'q2_IS_accts'] = deductions.loc[f[0],'q2_IS_accts']-1   
                if not(np.abs(round(ans2.loc['C22'][0],0))==np.abs(round(solution2.loc['C22'][0],0))):
                     deductions.loc[f[0],'q2_IS_accts'] = deductions.loc[f[0],'q2_IS_accts']-1   
                if not(np.abs(round(ans2.loc['C23'][0],0))==np.abs(round(solution2.loc['C23'][0],0))):
                     deductions.loc[f[0],'q2_IS_accts'] = deductions.loc[f[0],'q2_IS_accts']-1                        
                if not(np.abs(round(ans2.loc['C24'][0],0))==np.abs(round(solution2.loc['C24'][0],0))):
                     deductions.loc[f[0],'q2_IS_accts'] = deductions.loc[f[0],'q2_IS_accts']-1                     
                if not(np.abs(round(ans2.loc['C26'][0],0))==np.abs(round(ans2.loc['C25'][0]*(-0.21),0))):
                     deductions.loc[f[0],'q2_IS_accts'] = deductions.loc[f[0],'q2_IS_accts']-1                    
                if not(np.abs(round(ans2.loc['C28'][0],0))==np.abs(round(ans2.loc['C27'][0]*(-0.40),0))):
                     deductions.loc[f[0],'q2_IS_accts'] = deductions.loc[f[0],'q2_IS_accts']-1                    
             
 
                #Check first year balance sheet numbers
                if not(round(ans2.loc['C32'][0],0)==round(solution2.loc['C32'][0],0)):
                     deductions.loc[f[0],'q2_BS_accts'] = deductions.loc[f[0],'q2_BS_accts']-1                          
                if not(round(ans2.loc['C33'][0],0)==round(solution2.loc['C33'][0],0)):
                     deductions.loc[f[0],'q2_BS_accts'] = deductions.loc[f[0],'q2_BS_accts']-1   
                if not(round(ans2.loc['C35'][0],0)==round(solution2.loc['C35'][0],0)):
                     deductions.loc[f[0],'q2_BS_accts'] = deductions.loc[f[0],'q2_BS_accts']-1   
                if not(round(ans2.loc['C37'][0],0)==round(solution2.loc['C37'][0],0)):
                     deductions.loc[f[0],'q2_BS_accts'] = deductions.loc[f[0],'q2_BS_accts']-1   
                if not(round(ans2.loc['C40'][0],0)==round(solution2.loc['C40'][0],0)):
                     deductions.loc[f[0],'q2_BS_accts'] = deductions.loc[f[0],'q2_BS_accts']-1                        
                if not(round(ans2.loc['C41'][0],0)==round(solution2.loc['C41'][0],0)):
                     deductions.loc[f[0],'q2_BS_accts'] = deductions.loc[f[0],'q2_BS_accts']-1                          
                if not(round(ans2.loc['C43'][0],0)==round(solution2.loc['C43'][0],0)):
                     deductions.loc[f[0],'q2_BS_accts'] = deductions.loc[f[0],'q2_BS_accts']-1                                               
                if not(round(ans2.loc['C44'][0],0)==round(ans2.loc['C29'][0]+17000,0)):
                     deductions.loc[f[0],'q2_BS_accts'] = deductions.loc[f[0],'q2_BS_accts']-1  
                         
                #Check repurchase number correct
                     
 
            #Check data table numbers consistent with model above
            if not(np.abs(round(ans2.loc['C59'][0],0))==np.abs(round(ans2.loc['B49'][0],0)) and \
                   np.abs(round(ans2.loc['C53'][0],0))> np.abs(round(ans2.loc['C63'][0],0))):
                 deductions.loc[f[0],'q2_datatable'] = deductions.loc[f[0],'q2_datatable']-5   

            #Row error            
            if ans2.loc['B63'][0]!=solution2.loc['B63'][0]:
                print('Q2 row problem')
                deductions.loc[f[0],'q2_err']=1
                
            deductions.loc[f[0],'q2_total'] = sum(deductions.loc[f[0],'q2_repurchase':'q2_datatable'])
    except:
        print('Q2 Grade error')
        deductions.loc[f[0],'q2_err']=1   
      


    
    # Question 3 - max deduction is 12 points:
    cells_to_pull = cells_to_pull3
    ans3 = pd.DataFrame(index=cells_to_pull, columns=['q3'])
    try:
        for i in enumerate(cells_to_pull):
            col_to_pull =str(cells_to_pull[i[0]][0])
            row_to_pull =int(cells_to_pull[i[0]][1:])-1 
            ans3.iloc[i[0],0]=pd.read_excel(f[1],sheet_name='Q3', usecols=col_to_pull,skiprows=row_to_pull,nrows=1,header=None).iloc[0,0]   
    except:
        print('Q3 Read error')
        deductions.loc[f[0],'q3_err']=1 
  
    try:
        if deductions.loc[f[0],'q3_err']==0:
            
            if ans3.loc['C10'][0]>1:
                ans3.loc['C10'][0]=ans3.loc['C10'][0]/100
            
            #Q3a
            if not(round(ans3.loc['C10'][0],4)==round(solution3.loc['C10'][0],4)):
                    deductions.loc[f[0],'q3a'] = deductions.loc[f[0],'q3a']-4
            #Q3b
            if not(round(ans3.loc['C15'][0],2)==round(solution3.loc['C15'][0],2)):
                    deductions.loc[f[0],'q3b'] = deductions.loc[f[0],'q3b']-4
            #Q3c
            if not(round(ans3.loc['C20'][0],0)==round(solution3.loc['C20'][0],0)):
                    deductions.loc[f[0],'q3c'] = deductions.loc[f[0],'q3c']-4   
                    
            deductions.loc[f[0],'q3a_expl']=ans3.loc['B12'][0]                    
            deductions.loc[f[0],'q3b_expl']=ans3.loc['B17'][0]  
            deductions.loc[f[0],'q3c_expl']=ans3.loc['B22'][0]                    
            
            deductions.loc[f[0],'q3_total'] = sum(deductions.loc[f[0],'q3a':'q3c'])
                    
            #Row error
            if ans3.loc['B21'][0]!=solution3.loc['B21'][0]:
                print('Q3 row problem')
                deductions.loc[f[0],'q3_err']=1 
    except:
        print('Q3 Grade error')
        deductions.loc[f[0],'q3_err']=1   
  
```

#### Output grading file to excel and list results

```python
#Output to Excel  
deductions['sum']=100+deductions['q1_total']+deductions['q2_total']+deductions['q3_total']
deductions['filename']=file_list
deductions.to_excel(output_file)       
# deductions.tail(1)     
```

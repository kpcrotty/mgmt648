#### Packages and File Paths

```python
import os
import pandas as pd
import numpy as np
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

######################################
#File paths
######################################
PATH  = '/content/drive/MyDrive/Grading'

# solution_file = DROPBOX_PATH  + '/Week 4 Spreadsheet Assignment - SOLUTION.xlsx'
solution_file = 'https://www.dropbox.com/scl/fi/wg9sypqbheji9ny4ejopn/Week-4-Spreadsheet-Assignment-SOLUTION.xlsx?rlkey=ftmlms6mfpbv7gu5yk3dwtkly&st=r6q6zj29&dl=1'
ps_path       = PATH + '/ps4'
output_file   = PATH + '/grades_ps4.xlsx'

TOLERANCE = 0.10  #how far off the magnitudes can be for grading @risk output
```

#### Read in solutions

```python
######################################
#Read in solutions
######################################
cells_to_pull1 = ['C12','C14']
cells_to_pull = cells_to_pull1
solution1 = pd.DataFrame(index=cells_to_pull, columns=['q_AB'])
for i in enumerate(cells_to_pull):  
    col_to_pull =str(cells_to_pull[i[0]][0])
    row_to_pull =int(cells_to_pull[i[0]][1:])-1       
    solution1.iloc[i[0],0]=pd.read_excel(solution_file,sheet_name='Parts A & B', usecols=col_to_pull,skiprows=row_to_pull,nrows=1,header=None).iloc[0,0]



cells_to_pull2 = ['C8', 'C10']
cells_to_pull = cells_to_pull2
solution2 = pd.DataFrame(index=cells_to_pull, columns=['q_C'])
for i in enumerate(cells_to_pull):  
    col_to_pull =str(cells_to_pull[i[0]][0])
    row_to_pull =int(cells_to_pull[i[0]][1:])-1       
    solution2.iloc[i[0],0]=pd.read_excel(solution_file,sheet_name='Part C', usecols=col_to_pull,skiprows=row_to_pull,nrows=1,header=None).iloc[0,0]


cells_to_pull3 = ['C2', 'C3', 'C4','C5','B8']
cells_to_pull = cells_to_pull3
solution3 = pd.DataFrame(index=cells_to_pull, columns=['q_D'])
for i in enumerate(cells_to_pull):  
    col_to_pull =str(cells_to_pull[i[0]][0])
    row_to_pull =int(cells_to_pull[i[0]][1:])-1       
    solution3.iloc[i[0],0]=pd.read_excel(solution_file,sheet_name='Part D', usecols=col_to_pull,skiprows=row_to_pull,nrows=1,header=None).iloc[0,0]

```

#### Read in student solutions 

- should be downloaded and unzipped from canvas and place in Grading/ps4 folder of google drive

```python
######################################
# Read in student submissions
# Downloaded and unzipped from canvas    
######################################    
#Pull file names for the particular folder path
file_list = os.listdir(ps_path)
os.chdir(ps_path)
zeros = np.zeros(len(file_list), dtype=[('q_AB_err','f8'),('q_C_err','f8'), ('q_D_err','f8'),('q_AB','f8'),('q_C','f8'),('q_D','f8')])
deductions = pd.DataFrame(zeros)    


deductions['QAB feedback'] = 'QAB Deductions For: '
deductions['QC feedback'] = 'QC Deductions For: '
deductions['QD feedback'] = 'QD Deductions For: '



for f in enumerate(file_list):
# for f in enumerate([file_list[0]]):
    print(f[1])


    ##### Question A/B:
    cells_to_pull = cells_to_pull1
    ans1 = pd.DataFrame(index=cells_to_pull, columns=['q_AB'])
    try:
        for i in enumerate(cells_to_pull):
            col_to_pull =str(cells_to_pull[i[0]][0])
            row_to_pull =int(cells_to_pull[i[0]][1:])-1        
            #try: 
            ans1.iloc[i[0],0]=pd.read_excel(f[1],sheet_name='Parts A & B', usecols=col_to_pull,skiprows=row_to_pull,nrows=1,header=None).iloc[0,0]  
    except:
        print('Q_AB Read error')
        deductions.loc[f[0],'q_AB_err']=1  

    try:    
        if deductions.loc[f[0],'q_AB_err']==0:
            #Check Expected value of investment
            if not(ans1.loc['C12'][0]<(1+TOLERANCE)*(solution1.loc['C12'][0]) and ans1.loc['C12'][0]>(1-TOLERANCE)*(solution1.loc['C12'][0])):
                    deductions.loc[f[0],'q_AB'] = deductions.loc[f[0],'q_AB']-0.5
                    deductions.loc[f[0],'QAB feedback'] += '\nExpected value error; '


            #Check likelihood of >600K
            if not(ans1.loc['C14'][0]<(1+TOLERANCE)*(solution1.loc['C14'][0]) and ans1.loc['C14'][0]>(1-TOLERANCE)*(solution1.loc['C14'][0])):
                    deductions.loc[f[0],'q_AB'] = deductions.loc[f[0],'q_AB']-0.5
                    deductions.loc[f[0],'QAB feedback'] += '\nLikelihood < 600K error; '
            
    except:
        print('Q_AB Grade error')
        deductions.loc[f[0],'q_AB_err']=1   
        



    ##### Question C:
    cells_to_pull = cells_to_pull2
    ans2 = pd.DataFrame(index=cells_to_pull, columns=['q_C'])
    try:
        for i in enumerate(cells_to_pull):
            col_to_pull =str(cells_to_pull[i[0]][0])
            row_to_pull =int(cells_to_pull[i[0]][1:])-1        
            #try: 
            ans2.iloc[i[0],0]=pd.read_excel(f[1],sheet_name='Part C', usecols=col_to_pull,skiprows=row_to_pull,nrows=1,header=None).iloc[0,0]  
    except:
        print('Q_C Read error')
        deductions.loc[f[0],'q_C_err']=1  
 

    try:    
        if deductions.loc[f[0],'q_C_err']==0:
            #Check Expected value of investment
            if not(ans2.loc['C8'][0]<(1+TOLERANCE)*(solution2.loc['C8'][0]) and ans2.loc['C8'][0]>(1-TOLERANCE)*(solution2.loc['C8'][0])):
                    deductions.loc[f[0],'q_C'] = deductions.loc[f[0],'q_C']-0.5
                    deductions.loc[f[0],'QC feedback'] += '\nExpected value error; '


            #Check likelihood of >600K
            if not(ans2.loc['C10'][0]<(1+TOLERANCE)*(solution2.loc['C10'][0]) and ans2.loc['C10'][0]>(1-TOLERANCE)*(solution2.loc['C10'][0])):
                    deductions.loc[f[0],'q_C'] = deductions.loc[f[0],'q_C']-0.5
                    deductions.loc[f[0],'QC feedback'] += '\nLikelihood < 600K error; '
    except:
        print('Q_C Grade error')
        deductions.loc[f[0],'q_C_err']=1   


    ##### Question D:
    cells_to_pull = cells_to_pull3
    ans3 = pd.DataFrame(index=cells_to_pull, columns=['q_D'])
    try:
        for i in enumerate(cells_to_pull):
            col_to_pull =str(cells_to_pull[i[0]][0])
            row_to_pull =int(cells_to_pull[i[0]][1:])-1        
            #try: 
            #ans3? - i switched from ans2. Partc ->Part D
            ans3.iloc[i[0],0]=pd.read_excel(f[1],sheet_name='Part D', usecols=col_to_pull,skiprows=row_to_pull,nrows=1,header=None).iloc[0,0]  
    except:
        print('Q_D Read error')
        deductions.loc[f[0],'q_D_err']=1  
 

    try:    
        if deductions.loc[f[0],'q_D_err']==0:
            if round(ans3.loc['C2'][0],3)== 0.018 and round(ans3.loc['C4'][0],3)==0.082:
                print('Added zero as initial lagged value')
                deductions.loc[f[0],'q_D'] = -0.5    
                deductions.loc[f[0],'QD feedback'] += '\nAdded zero as initial lagged value; '
            else:
                if not(round(ans3.loc['C2'][0],3)==round(solution3.loc['C2'][0],3)):
                    deductions.loc[f[0],'q_D'] = deductions.loc[f[0],'q_D']-0.5  
                    deductions.loc[f[0],'QD feedback'] += '\nrho coefficient error; '
                if not(round(ans3.loc['C3'][0],3)==round(solution3.loc['C3'][0],3)):
                    deductions.loc[f[0],'q_D'] = deductions.loc[f[0],'q_D']-0.5 
                    deductions.loc[f[0],'QD feedback'] += '\nrho t-stat error; '
                if not(round(ans3.loc['C4'][0],3)==round(solution3.loc['C4'][0],3)):
                    deductions.loc[f[0],'q_D'] = deductions.loc[f[0],'q_D']-0.5 
                    deductions.loc[f[0],'QD feedback'] += '\nIntercept coefficient error; '
                if not(round(ans3.loc['C5'][0],3)==round(solution3.loc['C5'][0],3)):
                    deductions.loc[f[0],'q_D'] = deductions.loc[f[0],'q_D']-0.5  
                    deductions.loc[f[0],'QD feedback'] += '\nIntercept t-stat error; '
                           
        
    except:
        print('Q_D Grade error')
        deductions.loc[f[0],'q_D_err']=1  
    
deductions['QAB feedback'] = np.where(deductions['QAB feedback'] == 'QAB Deductions For: ' , '', deductions['QAB feedback'])
deductions['QC feedback'] = np.where(deductions['QC feedback'] == 'QC Deductions For: ' , '', deductions['QC feedback'])
deductions['QD feedback'] = np.where(deductions['QD feedback'] == 'QD Deductions For: ' , '', deductions['QD feedback'])
  
```

#### Output grading file to excel and list results

```python
# Output to Excel  
deductions['sum']=10+deductions['q_AB']+deductions['q_C']+deductions['q_D']
deductions['filename']=file_list
deductions.to_excel(output_file)    
```

#### Packages and File Paths

```python
import os
import pandas as pd
import numpy as np
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

######################################
#File paths
######################################
PATH  = '/content/drive/MyDrive/Grading'

# solution_file = PATH + '/Week 2 Spreadsheet Assignment - SOLUTION.xlsx'
solution_file = 'https://www.dropbox.com/scl/fi/4x3iurd8mm784rplhk3rw/Week-2-Spreadsheet-Assignment-SOLUTION.xlsx?rlkey=4uox89z64dh9mh7nbti35a2ox&st=3sk0s03r&dl=1'
ps_path       = PATH + '/ps2'
output_file   = PATH + '/grades_ps2.xlsx'

TOLERANCE = 0.125  #how far off the magnitudes can be for pro-forma entries
```

#### Read in solutions

```python
######################################
#Read in solutions
######################################
cells_to_pull1 = ['C42','C43', 'C44','C45', 'C46','C47', 'C48','C49', 
                  'F42','F43', 'F44','F45', 'F46','F47', 'F48','F49', 
                  'D49','E49',
                  'F50','B52','B53','B54', 'A54']
cells_to_pull = cells_to_pull1
solution1 = pd.DataFrame(index=cells_to_pull, columns=['q1'])
for i in enumerate(cells_to_pull):  
    col_to_pull =str(cells_to_pull[i[0]][0])
    row_to_pull =int(cells_to_pull[i[0]][1:])-1       
    solution1.iloc[i[0],0]=pd.read_excel(solution_file,sheet_name='A. FCF and Valuation', usecols=col_to_pull,skiprows=row_to_pull,nrows=1,header=None).iloc[0,0]


IS_accts  = ['G' + str(i) for i in np.arange(15,25)]
BS_accts1 = ['G28','G29'] + ['G' + str(i) for i in np.arange(31,34)]
BS_accts2 = ['G' + str(i) for i in np.arange(36,41)]
ASSETS    = ['G34','H34','I34','J34','K34']
LIABSEQ   = ['G41','H41','I41','J41','K41']
cells_to_pull2 = IS_accts + BS_accts1 + BS_accts2 + ASSETS + LIABSEQ + ['A41']
cells_to_pull = cells_to_pull2
solution2 = pd.DataFrame(index=cells_to_pull, columns=['q2'])
for i in enumerate(cells_to_pull):  
    col_to_pull =str(cells_to_pull[i[0]][0])
    row_to_pull =int(cells_to_pull[i[0]][1:])-1       
    solution2.iloc[i[0],0]=pd.read_excel(solution_file,sheet_name='B. Pro-forma (Debt Plug)', usecols=col_to_pull,skiprows=row_to_pull,nrows=1,header=None).iloc[0,0]
```

#### Read in student solutions 

- should be downloaded and unzipped from canvas and place in Grading/ps2 folder of google drive

```python
######################################
# Read in student submissions
# Downloaded and unzipped from canvas    
######################################    


#Pull file names for the particular folder path
file_list = os.listdir(ps_path)
os.chdir(ps_path)
zeros = np.zeros(len(file_list), dtype=[('q1_err','f8'),('q2_err','f8'),     ('q1_FCFs','f8'),('q1_TV','f8'),('q1_EV','f8'),('q1_total','f8'),    \
                                        ('q2_IS_accts','f8'),('q2_BS_accts','f8'),('q2_bs_tie','f8')])
deductions = pd.DataFrame(zeros)    

deductions['FCF feedback'] = 'FCF Deductions For: '
deductions['Pro-forma feedback'] = 'Pro-forma Deductions For: '


for f in enumerate(file_list):
# for f in enumerate([file_list[-3]]):
    print(f[1])


    # Question 1 - max deduction is 21 points:
    cells_to_pull = cells_to_pull1
    ans1 = pd.DataFrame(index=cells_to_pull, columns=['q1'])
    try:
        for i in enumerate(cells_to_pull):
            col_to_pull =str(cells_to_pull[i[0]][0])
            row_to_pull =int(cells_to_pull[i[0]][1:])-1        
            try: 
                ans1.iloc[i[0],0]=pd.read_excel(f[1],sheet_name='A. FCF and Valuation', usecols=col_to_pull,skiprows=row_to_pull,nrows=1,header=None).iloc[0,0]  
            except ValueError:
                print('a first error')
                deductions.loc[f[0],'q1'] =-10
    except:
        print('Q1 Read error')
        deductions.loc[f[0],'q1_err']=1  
 

    try:    
        if deductions.loc[f[0],'q1_err']==0:
            #Check FCF in year 1
            if not(round(ans1.loc['C42'][0],0)==round(solution1.loc['C42'][0],0)):
                 deductions.loc[f[0],'q1_FCFs'] = deductions.loc[f[0],'q1_FCFs']-1                          
            if not(round(ans1.loc['C43'][0],0)==round(solution1.loc['C43'][0],0)):
                 deductions.loc[f[0],'q1_FCFs'] = deductions.loc[f[0],'q1_FCFs']-1   
            if not(np.abs(round(ans1.loc['C44'][0],0))==np.abs(round(solution1.loc['C44'][0],0))):
                 deductions.loc[f[0],'q1_FCFs'] = deductions.loc[f[0],'q1_FCFs']-1   
            if not(round(ans1.loc['C45'][0],0)==round(solution1.loc['C45'][0],0)):
                 deductions.loc[f[0],'q1_FCFs'] = deductions.loc[f[0],'q1_FCFs']-1   
            if not(round(ans1.loc['C46'][0],0)==round(solution1.loc['C46'][0],0)):
                 deductions.loc[f[0],'q1_FCFs'] = deductions.loc[f[0],'q1_FCFs']-1                        
            if not(np.abs(round(ans1.loc['C47'][0],0))==np.abs(round(solution1.loc['C47'][0],0))):
                 deductions.loc[f[0],'q1_FCFs'] = deductions.loc[f[0],'q1_FCFs']-1  
            if not(np.abs(round(ans1.loc['C48'][0],0))==np.abs(round(solution1.loc['C48'][0],0))):
                 deductions.loc[f[0],'q1_FCFs'] = deductions.loc[f[0],'q1_FCFs']-1  
            if not(round(ans1.loc['C49'][0],0)==round(solution1.loc['C49'][0],0)):
                 deductions.loc[f[0],'q1_FCFs'] = deductions.loc[f[0],'q1_FCFs']-1  
    
            
            #Fheck FCF in year 5
            if not(round(ans1.loc['F42'][0],0)==round(solution1.loc['F42'][0],0)):
                 deductions.loc[f[0],'q1_FCFs'] = deductions.loc[f[0],'q1_FCFs']-1                          
            if not(round(ans1.loc['F43'][0],0)==round(solution1.loc['F43'][0],0)):
                 deductions.loc[f[0],'q1_FCFs'] = deductions.loc[f[0],'q1_FCFs']-1   
            if not(np.abs(round(ans1.loc['F44'][0],0))==np.abs(round(solution1.loc['F44'][0],0))):
                 deductions.loc[f[0],'q1_FCFs'] = deductions.loc[f[0],'q1_FCFs']-1   
            if not(round(ans1.loc['F45'][0],0)==round(solution1.loc['F45'][0],0)):
                 deductions.loc[f[0],'q1_FCFs'] = deductions.loc[f[0],'q1_FCFs']-1   
            if not(round(ans1.loc['F46'][0],0)==round(solution1.loc['F46'][0],0)):
                 deductions.loc[f[0],'q1_FCFs'] = deductions.loc[f[0],'q1_FCFs']-1                        
            if not(np.abs(round(ans1.loc['F47'][0],0))==np.abs(round(solution1.loc['F47'][0],0))):
                 deductions.loc[f[0],'q1_FCFs'] = deductions.loc[f[0],'q1_FCFs']-1  
            if not(np.abs(round(ans1.loc['F48'][0],0))==np.abs(round(solution1.loc['F48'][0],0))):
                 deductions.loc[f[0],'q1_FCFs'] = deductions.loc[f[0],'q1_FCFs']-1  
            if not(round(ans1.loc['F49'][0],0)==round(solution1.loc['F49'][0],0)):
                 deductions.loc[f[0],'q1_FCFs'] = deductions.loc[f[0],'q1_FCFs']-1  
    
            #Cap deduction
            if deductions.loc[f[0],'q1_FCFs'] <= -1:
                deductions.loc[f[0],'q1_FCFs'] = -0.5
    
       
            #Check terminal value                   
            if not(round(ans1.loc['F50'][0],0)==round(ans1.loc['F49'][0]*(1.02)/(0.07-0.02),0)):
                 deductions.loc[f[0],'q1_TV'] = deductions.loc[f[0],'q1_TV']-0.5                    
                   
         
     
            #Check NPV
            npv = ans1.loc['C49'][0]/(1.07**1) + ans1.loc['D49'][0]/(1.07**2) + ans1.loc['E49'][0]/(1.07**3) + ans1.loc['F49'][0]/(1.07**4)
            npv_tv = ans1.loc['F50'][0]/(1.07**4)
            if not(round(ans1.loc['B52'][0],0)==round(npv,0)):
                 deductions.loc[f[0],'q1_EV'] = deductions.loc[f[0],'q1_EV']-1      
            if not(round(ans1.loc['B53'][0],0)==round(npv_tv,0)):
                 deductions.loc[f[0],'q1_EV'] = deductions.loc[f[0],'q1_EV']-1      
            if not(round(ans1.loc['B54'][0],0)==round(ans1.loc['B52'][0]+ans1.loc['B53'][0],0)):
                 deductions.loc[f[0],'q1_EV'] = deductions.loc[f[0],'q1_EV']-1  
            #Cap deduction
            if deductions.loc[f[0],'q1_EV'] <= -1:
                deductions.loc[f[0],'q1_EV'] = -0.5                   
    

            if deductions.loc[f[0],'q1_FCFs'] < 0:
               deductions.loc[f[0],'FCF feedback'] += '\nFCF error; '
            if deductions.loc[f[0],'q1_TV'] < 0:
               deductions.loc[f[0],'FCF feedback'] += '\nTerminal Value error; '
            if deductions.loc[f[0],'q1_EV'] < 0:
               deductions.loc[f[0],'FCF feedback'] += '\nEnterprise value error; '

            #Row error            
            if ans1.loc['A54'][0]!=solution1.loc['A54'][0]:
                print('Q1 row problem')
                deductions.loc[f[0],'q1_err']=1
                
            deductions.loc[f[0],'q1_total'] = sum(deductions.loc[f[0],'q1_FCFs':'q1_EV'])
    except:
        print('Q1 Grade error')
        deductions.loc[f[0],'q1_err']=1   



    # Question 2 (Pro-forma) - max deduction is X points:
    cells_to_pull = cells_to_pull2
    ans2 = pd.DataFrame(index=cells_to_pull, columns=['q2'])
    try:
        for i in enumerate(cells_to_pull):
            col_to_pull =str(cells_to_pull[i[0]][0])
            row_to_pull =int(cells_to_pull[i[0]][1:])-1        
            try: 
                ans2.iloc[i[0],0]=pd.read_excel(f[1],sheet_name='B. Pro-forma (Debt Plug)', usecols=col_to_pull,skiprows=row_to_pull,nrows=1,header=None).iloc[0,0]  
            except ValueError:
                print('a first error')
                deductions.loc[f[0],'q2'] =-10
    except:
        print('Q2 Read error')
        deductions.loc[f[0],'q2_err']=1  
 

    try:    
        if deductions.loc[f[0],'q2_err']==0:
            #Check balance sheet ties
            if not(round(ans2.loc['G34'][0],0)==round(ans2.loc['G41'][0],0) and \
                   round(ans2.loc['H34'][0],0)==round(ans2.loc['H41'][0],0) and \
                   round(ans2.loc['I34'][0],0)==round(ans2.loc['I41'][0],0) and \
                   round(ans2.loc['J34'][0],0)==round(ans2.loc['J41'][0],0) and \
                   round(ans2.loc['K34'][0],0)==round(ans2.loc['K41'][0],0) ):
                 deductions.loc[f[0],'q2_bs_tie'] = deductions.loc[f[0],'q2_bs_tie']-3              
            
            
            #Check first year income statement numbers
            for cell in IS_accts:
                if not(np.abs(ans2.loc[cell][0])<(1+TOLERANCE)*(np.abs(solution2.loc[cell][0])) and np.abs(ans2.loc[cell][0])>(1-TOLERANCE)*(np.abs(solution2.loc[cell][0]))):
                        print('Deduction for cell#: {:}'.format(cell))
                        deductions.loc[f[0],'q2_IS_accts'] = deductions.loc[f[0],'q2_IS_accts']-0.5       
                        
             #Check first year balance sheet numbers
            for cell in BS_accts1:
                if not(np.abs(ans2.loc[cell][0])<(1+TOLERANCE)*(np.abs(solution2.loc[cell][0])) and np.abs(ans2.loc[cell][0])>(1-TOLERANCE)*(np.abs(solution2.loc[cell][0]))):
                        print('Deduction for cell#: {:}'.format(cell))
                        deductions.loc[f[0],'q2_BS_accts'] = deductions.loc[f[0],'q2_BS_accts']-0.5      
            for cell in BS_accts2:
                if not(np.abs(ans2.loc[cell][0])<(1+TOLERANCE)*(np.abs(solution2.loc[cell][0])) and np.abs(ans2.loc[cell][0])>(1-TOLERANCE)*(np.abs(solution2.loc[cell][0]))):
                        print('Deduction for cell#: {:}'.format(cell))
                        deductions.loc[f[0],'q2_BS_accts'] = deductions.loc[f[0],'q2_BS_accts']-0.5      

            #Row error            
            if ans2.loc['A41'][0]!=solution2.loc['A41'][0]:
                print('Q2 row problem')
                deductions.loc[f[0],'q2_err']=1
 
               
            if deductions.loc[f[0],'q2_IS_accts'] < 0:
               deductions.loc[f[0],'Pro-forma feedback'] += '\nIncome statement error; '
            if deductions.loc[f[0],'q2_BS_accts'] < 0:
               deductions.loc[f[0],'Pro-forma feedback'] += '\nBalance sheet error; '




            deductions.loc[f[0],'q2_total'] = sum(deductions.loc[f[0],'q2_IS_accts':'q2_bs_tie'])
            #Cap deduction at -2
            if deductions.loc[f[0],'q2_total']<-3:
                deductions.loc[f[0],'q2_total']=-3                    

    except:
        print('Q2 Grade error')
        deductions.loc[f[0],'q2_err']=1   


deductions['FCF feedback'] = np.where(deductions['FCF feedback'] == 'FCF Deductions For: ' , '', deductions['FCF feedback'])
deductions['Pro-forma feedback'] = np.where(deductions['Pro-forma feedback'] == 'Pro-forma Deductions For: ' , '', deductions['Pro-forma feedback'])        
```

#### Output grading file to excel and list results

```python
#Output to Excel  
deductions['sum']=10+deductions['q1_total']+deductions['q2_total']
deductions['filename']=file_list
deductions.to_excel(output_file)    
```
